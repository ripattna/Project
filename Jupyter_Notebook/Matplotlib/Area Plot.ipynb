```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5

# Add figure
fig = plt.figure(figsize=(10,6))
#Sub-plotting
ax1 = plt.subplot()

# Creating the area plot 
plt.stackplot(x,y, color='g', alpha=0.9)
plt.plot(x,y, color = 'maroon')
plt.title("Area Plot Sample")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.show()
```
