```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Defining the dataset
x = np.arange(0,10,0.1)
y = 2*x + 5

# Plotting the datapoints
plt.plot(x,y)
plt.show()
```

# Customize the Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5

# Changing the figure
fig = plt.figure(figsize=(10,5))

# Plotting the datapoints
plt.plot(x,y,linewidth = 2.0, linestyle = ":", color = 'r', alpha = 0.5, marker = 'o')
plt.title("Line Plot Sample")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
y1= 5*x + 6

# Changing the figure
fig = plt.figure(figsize=(10,5))

# Plotting the datapoints
plt.plot(x,y,linewidth = 2.0, linestyle = ":", color = 'blue', alpha = 0.5, marker = 'o')
plt.plot(x,y1,linewidth = 2.0, linestyle = ":", color = 'orange', alpha = 0.8, marker = 'X')
plt.title("Line Plot Sample")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','my_line'],loc='best')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Defining the dataset
x = np.arange(0,10,1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 5*x + 6

#Creating two or more plots in the fig
plt.subplot(1,3,1)

#Plotting the points
plt.plot(x,y1)
plt.title('Graph-1')

plt.subplot(1,3,2)
plt.plot(x,y2)
plt.title('Graph-2')

plt.subplot(1,3,3)
plt.plot(x,y3)
plt.title('Graph-3')
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Defining the dataset
x = np.arange(0,10,1)
y1 = 2*x + 5
y2 = 3*x + 10
y3 = 5*x + 6

#Creating two or more plots in the fig
plt.subplot(221)
plt.plot(x,y1)
plt.title('Graph-1')

plt.subplot(222)
plt.plot(x,y2)
plt.title('Graph-2')

plt.subplot(223)
plt.plot(x,y3)
plt.title('Graph-3')
```

```python
import matplotlib.pyplot as plt

#Creating the datasets
data = {'Apple':20, 'Mango':15, 'Lemon':30, 'Orange':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize=(10,5))

# Creatingt the bar plot
plt.plot(names,values)
plt.title("Line Plot Sample")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```
