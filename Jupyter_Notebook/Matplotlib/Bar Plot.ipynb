```python
import matplotlib.pyplot as plt

#Creating the datasets
data = {'Apple':20, 'Mango':15, 'Lemon':30, 'Orange':10}
names = list(data.keys())
values = list(data.values())

# Add figure
fig = plt.figure(figsize=(10,5))

# Creating the bar plot
# plt.bar(names,values, color="orange")
plt.style.use('dark_background')
plt.barh(names,values, color="brown")
# Customizing title
plt.title("Bar Plot Demo")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

```python
# import library
import matplotlib.pyplot as plt
# %matplotlib inline

#Creating the datasets
data = {'Apple':10, 'Mango':15, 'Lemon':5, 'Orange':20}
names = list(data.keys())
values = list(data.values())

# Add figure
fig = plt.figure(figsize=(10,5))
#Sub-plotting
ax1 = plt.subplot()

# Creating the bar plot
ax1.barh(names, values, align='center', color="orange")
# Customizing title
plt.title("Bar Plot Sample")
# Customization-x axis label, y-axis label
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
# Customization-add grid
plt.grid(True)
# Save the plot
# plt.savefig('HorizontalBarPlot.png')
# Save Plot
plt.show()
```
