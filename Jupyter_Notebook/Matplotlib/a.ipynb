# Scatter Plot

```python
import matplotlib.pyplot as plt

# Creating the datasets
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# Creating the scatter plot
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating the datasets
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# Creating the scatter plot
plt.scatter(a,b)
plt.scatter(a,x)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1','my_line'],loc='best')
plt.show()
```
