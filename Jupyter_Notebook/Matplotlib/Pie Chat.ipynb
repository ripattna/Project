```python
import matplotlib.pyplot as plt

# Prepare the data
labels = ['Cats', 'Horse', 'Dogs', 'Lion']
sizes = [15,30,45,10]

# Plot-sezes, labels, autopercentage, shodow, start-angle=90
plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow = True, startangle=90)
# Customization-title
plt.title('Pie Chat')
#save the plot
#plt.savefig('PieChat.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Prepare the data
labels = ['Dog', 'Cat', 'Wolf', 'Lion']
sizes = [50,45,60,80]

# Add figure
fig =plt.figure(figsize=(10,9))
# Add axes
ax1 = plt.subplot()

# Plot-sezes, labels, autopercentage, shodow, start-angle=90
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow = True, startangle=90, explode = (0.1,0.1,0,0))

# Customization-title
plt.title('Pie Chat')
#save the plot
#plt.savefig('PieChat.png')
plt.show()
```
