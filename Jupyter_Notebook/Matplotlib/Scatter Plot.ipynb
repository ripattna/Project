# Scatter Plot

```python
import matplotlib.pyplot as plt

# Creating the datasets
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# Creating the scatter plot
plt.scatter(a,b)
plt.scatter(a,x)
plt.show()
```

# Customizing the Scatter Plot

```python
import matplotlib.pyplot as plt

# Creating the datasets
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# Customization-color,size,edgecolor,marker,alpha
plt.scatter(a,b, color='g', s=200, edgecolors='y', marker='o', alpha=0.5)
plt.scatter(a,x, c='maroon', s=400, edgecolors='y', marker='4', alpha=1)
plt.legend(['b','x'], loc='best')
# Customization title
plt.title("Scatter Plot Sample")
# Customization-x axis label, y-axis label
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
# Customization-add grid
plt.grid(True)
# plt.savefig('Scatter_Plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creating the data
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

# Add figure
fig = plt.figure(figsize=(10,5))
# Add axes
ax1 = plt.subplot()
# Customization-color,size,edgecolor,marker,alpha
ax1.scatter(a,b, color='g', s=200, edgecolors='y', marker='o', alpha=0.5)
ax1.scatter(a,x, c='maroon', s=400, edgecolors='y', marker='4', alpha=1)
plt.legend(['b','x'], loc='best')
# Customization title
plt.title("Scatter Plot Sample")
# Customization-x axis label, y-axis label
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
# Customization-add grid
plt.grid(True)
# plt.savefig('Scatter_Plot.png')
plt.show()
```
