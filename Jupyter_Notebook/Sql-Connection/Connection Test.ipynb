```python
import mysql.connector
from mysql.connector import errorcode

try:
    cnx = mysql.connector.connect(host="localhost", user='root', password='root', use_pure=True)

except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Something is wrong with your user name or password")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("Database does not exist")
    else:
        print(err)

else:
    print("Connection is established!")
    # cnx.close()

my_cursor = cnx.cursor()
my_cursor.execute("show databases")
for x in my_cursor:
    print(x)
```

```python
def reverse(s):
    str1 = ""
    for i in s:
        str = i + str1
    return str1


s = "Geeksforgeeks"

print("The original string  is : ", end="")
print(s)

print("The reversed string(using loops) is : ", end="")
print(reverse(s))
```

```python
import mysql.connector
from mysql.connector import Error

try:
    connection = mysql.connector.connect(host='localhost',database='demo',user='root',password='root')
    
    if connection.is_connected():
        db_Info = connection.get_server_info()
        print("Connected to MySQL Server version:", db_Info)
        cursor = connection.cursor()
        cursor.execute("select database();")
        #cursor.execute("select * from demo.employee limit 5;")
        record = cursor.fetchone()
        print("You're connected to database: ", record)

except Error as e:
    print("Error while connecting to MySQL", e)
    
finally:
    if (connection.is_connected()):
        cursor.close()
        connection.close()
        print("MySQL connection is closed!")
```

```python
import mysql.connector
from mysql.connector import Error

try:
    connection = mysql.connector.connect(host='localhost',
                                         database='demo',
                                         user='root',
                                         password='root')

    mySql_Create_Table_Query = """CREATE TABLE employee_new ( 
                             Id int(11) NOT NULL,
                             Name varchar(250) NOT NULL,
							 Gender varchar(250) NOT NULL,
                             Salary float NOT NULL,
                             PRIMARY KEY (Id)) """

    cursor = connection.cursor()
    result = cursor.execute(mySql_Create_Table_Query)
    print("Employee Table created successfully ")

except mysql.connector.Error as error:
    print("Failed to create table in MySQL: {}".format(error))
finally:
    if (connection.is_connected()):
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```
